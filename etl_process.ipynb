```python
#separamos a la columna 'duration' de nuestro df en dos columnas

duration = df['duration'].str.split(' ', expand=True)
```

```python
# Cambio el tipo de dato de la columna duration_time para poder cuantificarla y lo asignamos a una Serie

duration_time = pd.to_numeric(duration.duration_time)
```

```python
duration['min/season'].value_counts()
```

```python
#combinamos las dos categorías Season y Seasons en una sola

min_season = duration['min/season'].replace({'Season':'season', 'Seasons':'season'})
```

```python
#combinamos ambas series 'duration' y 'min_season'
duration = pd.concat([duration_time, min_season], axis=1)
```

```python
lista_unicos = []
lista_repeticiones = []
def valor_modal(lista):
    
    if len(lista) == 0:
        return None
    for elemento in lista:
        if elemento in lista_unicos:
            i = lista_unicos.index(elemento)
            lista_repeticiones[i] += 1
        else:
            lista_unicos.append(elemento)
            lista_repeticiones.append(1)
    moda = lista_unicos[0]
    maximo = lista_repeticiones[0]
    for i, elemento in enumerate(lista_unicos):
        if lista_repeticiones[i] > maximo:
            moda = lista_unicos[i]
            maximo = lista_repeticiones[i]
    return moda, maximo, len(lista_unicos), len(lista_repeticiones)
```
